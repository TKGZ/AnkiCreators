# Creating Anki Deck from Unity Glossary


```
from bs4 import BeautifulSoup
import pandas as pd
import numpy as np
import requests
r = requests.get('https://docs.unity3d.com/Manual/Glossary.html')
c = r.content
soup = BeautifulSoup(c, 'html.parser')
```

#Saving Definitions to CSV


```
results = []
g = soup.find("div", {"class":"Glossary"})
while (g != None):
  subject = g['title']
  w = g.find("h4")
  while (w != None):
    d = w.find_next_sibling("p").getText().split("More info")[0]
    
    results.append([w.getText()[:-1], d, subject])

    w = w.find_next_sibling("h4")
  g = g.find_next_sibling("div", {"class":"Glossary"})

print("printing the df")
df = pd.DataFrame(results, columns=['word', 'definition', 'type'])
```
